# BAGGING


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#load the Iris dataset
iris = load_iris()
X,Y = iris.data, iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)#traing and testing split
```

```python
base_model = DecisionTreeClassifier()#base model
bag_model = BaggingClassifier(estimator=base_model, n_estimators=10, random_state=42)#creating basemodel (estimator = base model)
```

```python
base_model.fit(X_train, Y_train)#traing both model
bag_model.fit(X_train, Y_train)
```

```python
y_pred_base = base_model.predict(X_test)#predication
y_pred_bag = bag_model.predict(X_test)
```

```python
print("Accuracy of single decision tree:", accuracy_score(Y_test, y_pred_base))
print("Accuracy of bagged decision tree:", accuracy_score(Y_test, y_pred_bag))
```

# GRADIENT BOOST

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#dataset
iris =load_iris()
X,y= iris.data, iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=42)#Spliting into x and y
```

```python
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)#Creating model Boosting mode #
```

```python
gb_model.fit(X_train, Y_train)#train fit the model
```

```python
#predictions
y_pred = gb_model.predict(X_test)
```

```python
#evaaluate performance
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion MMstric:")
print(confusion_matrix(Y_test,y_pred))
print("Classification Report:")
print(classification_report(Y_test,y_pred))
```

```python
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#load the Iris dataset
iris = load_iris()
X,Y = iris.data, iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)#traing and testing split
```

```python
#create xgboost model
xgb_model = xgb.XGBClassifier(
    n_estimators=100,#number of trees
    learning_rate=0.1,#step size shrinkage
    max_depth=3,#max depth of a tree
    subsample=0.8,#fraction of training samples to use
    colsample_bytree=0.8,#fraction of features
    random_state=42,
    eval_metric='mlogloss'#multiclass log loss metric
)
```

```python
xgb_model.fit(X_train, Y_train)#train
```

```python
#predictions
y_pred = xgb_model.predict(X_test)

#evaluate performance

print("Accuracy of XGBoost model:", accuracy_score(Y_test, y_pred))
print("Confusion MMstric:")
print(confusion_matrix(Y_test,y_pred))
print("Classification Report:")
print(classification_report(Y_test,y_pred))
```
