```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
X = np.array([
    [2,9],
    [1,5],
    [3,6],
    [4,8],
    [6,9],
    [5,5],
    [7,3]],dtype=float)
```

```python
y = np.array([
    [0],
    [0],
    [0],
    [1],
    [1],
    [0],
    [1]
],dtype=float)
```

```python
X = X/np.max(X, axis=0)
```

```python
model =keras.Sequential([
    layers.Dense(4,input_shape=(2,),activation='relu'),
    layers.Dense( 1,activation='sigmoid')])
```

```python
 model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=200,verbose=0)
```

```python
test_data = np.array([[3,9]])/np.amax(X,axis=0)
prediction = model.predict(test_data)
print("Predicted output 1=pass 0=false):",prediction)
if prediction >=0.5:
  print("The student is likely to Pass")
else:
  print("the student is likely to Fail")
```
