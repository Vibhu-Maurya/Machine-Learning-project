```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data
ManPower = np.array([20, 40, 50, 60, 70, 80,90,100]).reshape(-1, 1)
Day_Require = np.array([1,2,3,4,5,6,7,8])
```

```python
# Fit model
model = LinearRegression()
model.fit(ManPower, Day_Require)
```

```python
m = model.coef_[0]
b = model.intercept_
print(f"slope (m):{m}")
print(f"interception (b) :{b}")
```

```python
Day_Require_pred = model.predict(ManPower)

plt.scatter(ManPower, Day_Require, color="blue", label="Actual Data")
plt.plot(ManPower, Day_Require_pred, color="red", label="Regression Line (through origin)")
plt.xlabel("ManPower")
plt.ylabel("Day Require")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Upload your CSV file
# Example: CSV contains columns "ManPower" and "Day_Require"
from google.colab import files
uploaded = files.upload()
file_path = "Around_the_World_in_Eighty_Days.csv"  # replace with your file name
df = pd.read_csv(file_path)

# 2. Extract features and target
X = df[['ManPower']]  # Independent variable(s)
y = df['Day_Require']  # Dependent variable

# 3. Train the model
model = LinearRegression()
model.fit(X, y)

# 4. Predictions
y_pred = model.predict(X)

# 5. Print model parameters
print(f"Slope (m): {model.coef_[0]:.4f}")
print(f"Intercept (b): {model.intercept_:.4f}")
print(f"Equation: Day_Require = {model.coef_[0]:.4f} * ManPower + {model.intercept_:.4f}")

# 6. Predict for a new value
new_value = np.array([[75]])
predicted = model.predict(new_value)
print(f"Predicted days for manpower {new_value[0][0]}: {predicted[0]:.2f}")

# 7. Plot
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.xlabel("ManPower")
plt.ylabel("Day Require")
plt.legend()
plt.show()
```
